
# **PIVOTING DATA FRAMES**

*   Columnlar valuelara, valuelar da columnlara dönüştürülür.
*   Dataframe'e farklı bir bakış açısı getirir.







```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
#bir dictionary oluşturuldu.
df = pd.DataFrame(dic)  #bu dictionary dataframe'e dönüştürüldü.
df
```

```python
df.pivot(index="treatment",columns = "gender",values="response")  #index treatment sütunu yapıldı.
                                                                  #columnlar gender sütundaki verilerden (male,female) oluşur.
                                                                  #valuelar response sütunu yani tedaviye yanıt verme durumudur.
```

# **STACKING and UNSTACKING DATAFRAME**
*   Birden çok indeks olduğunda o indexlerle ilgilenilmesi gerekebilir.
*   **swaplevel**: inner ve outer indekslerin yerini değiştirmektir.
*   **level:** outer indeks level=0, inner indeks level=1'dir.






```python
df1 = df.set_index(["treatment","gender"])  #treatment outer indeks, gender inner indekstir.
df1
```

```python
df1.unstack(level=0)               #treatment sütunu indekslikten çıkarılmıştır, sadece gender indekstir.
```

```python
df1.unstack(level=1)             #gender sütunu indekslikten çıkarılmıştır, sadece treatment indekstir.
```

```python
df2 = df1.swaplevel(0,1)     #inner ve outer indekslerin yeri değiştirilir.
df2
```

# **CATEGORICALS AND GROUPBY**

```python
df
```

```python
df.groupby("treatment").mean()    #treatment sütunundaki verilere göre gruplanır ve sütunların ortalaması alınır.
                                  #response sütununda treatment B olan veriler 5 ve 9'dur.
                                  #(5+9)/2=7 treatment B'nin response değeri 7'dir.
```

```python
df.groupby("treatment").age.max() #treatment sütununa göre gruplanır.
                                  #age sütununun max değerleri alınır.
```

```python
df.groupby("treatment")[["age","response"]].min() #treatmenta göre gruplanır.
                                                  #age ve respons sütunlarının min değerleri alınır.
```

```python
df.info()   #treatment ve gender sütunlarının datatype'ı objecttir.
            #String değer barındırır.
            #Bu sütunların türü category yapılabilir.
            #Çünkü categorical data hafızayı daha az kullanılır, group by gibi operationları hızlandırır.
```
