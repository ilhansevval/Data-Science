# **PANDAS**


*   Pandas kütüphanesi dataframeler için oluşturulmuş bir kütüphanedir.
*   Dataframeler datayı içinde depoladığımız yapılardır. 
*   Pandas kütüphanesi ile dosyalar arası geçiş kolaylaşır.
*   csv ve text dosyaları üzerinde işlem yapılıp, sonuçlar tekrardan kaydedilebilir.
*   csv (comma seperated values) bu dosyalarda veriler virgüller ile ayrılmıştır. Pandas kütüphanesi ile bu veriler dataframe'e çevrilir.
*   Pandas'ta iki çeşit datatype bulunur: series ve dataframes
*   series tek boyutlu vektörler, dataframe'ler çok boyutludur.










```python
series = data['Defense']        #tek boyutlu vektör
print(type(series))
data_frame = data[['Defense']]  #dataframe oluştururken seçilen feature bir liste içerisine yazılır.
print(type(data_frame))
```

```python
# Comparison operator
print(3 > 2)   #Comparison operatorları >, <, ==, !=, >=, <=
print(3!=2)
# Boolean operators
print(True and False) #boolean operatorları and, or, not operatorlarıdır, boolean değerleri True, False'tur.
print(True or False)
```

```python
data['Defense']>200 #değerler boolean olarak true/false şeklinde döndürülür, şartı sağlayan değerler True'dur.
```

```python
x = data['Defense']>200     # Defense'i 200'den büyük olan 3 pokemon vardır.
data[x]                     #Defense'i 200'den büyük olanlar x filtresine aktarılır.
                            #x filtresi datanın içine yazıldığında defense değerleri false olanları ignore eder, true'ları döndürür.
```

```python
#Filtering pandas with logical_and
data[np.logical_and(data['Defense']>200, data['Attack']>100 )] #logical_and() methoduyla iki durumun da değerlerinin true olsuğu sonuçlar verilir.
```

```python
data[(data['Defense']>200) & (data['Attack']>100)] #numpy kütüphanesinde bulunan logical_and() methodunu kullanılmadan da aynı işlem & operatörüyle gerçekleştirilebilir.
```
