```python
import warnings
warnings.filterwarnings("ignore")

data2 = data.head() #dataframe'deki ilk 5 satır data2 değişkenine atandı.
date_list = ["1992-01-10","1992-02-10","1992-03-10","1993-03-15","1993-03-16"] #time listesi oluşturuldu.
datetime_object = pd.to_datetime(date_list)  #listedeki verilerin türü datetime'a çevrildi.
data2["date"] = datetime_object  #datetime'a çevrilmiş veriler date adlı sütuna atandı.

data2= data2.set_index("date")   #set_index methodu ile date sütunundaki verilerin index olması sağlandı.
data2 
```

loc methodu ile belli indeksteki datalar çekilir.

Burada indeks tarih olduğu için tarih yazılır.

```python
print(data2.loc["1993-03-16"])
```

loc methodu pandasta (inclusive:inclusive)'dir.

```python
print(data2.loc["1992-03-10":"1993-03-16"]) 
```

# **RESAMPLING PANDAS TIME SERIES**
*   **Resampling:** farklı zaman aralıklarındaki istatiksel methoddur.
*   Resample methodu ile data içerisinden yeni samplelar elde edilir.
*   Sıklığı belirlemek için "M" = month ya da "A" = year gibi string değerler gerekir.



```python
data2.resample("A").mean()  #data2 içerisindeki yıllara göre sütunların ortalama değerleri bulunur.
                            #data2 içerisinde 2 yıl var, bu yüzden 2 adet sonuç döner.
                            #HP sütunu can değerleridir
                            #1992 yılında 45,60,80 olmak üzere 3 tane can değeri var.
                            #Bunların ortalaması (45+60+80)/3=61.6 bulunur.
                            #Tüm sütunlarda aynı işlem uygulanır.
```

```python
data2.resample("M").mean() #Sütunların aylara göre ortalaması bulunur.
                           #1 yılda 12 ay vardır.
                           #1992 yılı için ilk 3 ay resample yapılır.
                           #3. aydan sonraki aylar veri setinde olmadığı için NaN değer döndürülür.
                           #dataframe'de 1993 yılı için sadece 3. ayda veri var.
                           #Bu yüzden ilk iki ayda NaN değer döndürülür.
                           #1992 yılındaki resample edilmiş verilerin dataframe'deki verilerle aynı olduğu görülür.
                           #Bunun sebebi 1.,2. ve 3. aylara ait tek bir satır olmasıdır.
                           #1993 yılında 4.ayı yazdırmamasının sebebi resample methodu en erken tarih ile en geç tarih arasını doldurur.
        #1992 yılı 1.aydan başladığı için resample methodu 1. aydan başlar, 1993 yılı 3.ayda bittiği için en son 1993 yılı 3.ayda biter.
```

interpolate methoduyla sütunlardaki missing değerler doldurulur.

Sayısal değer barındıran sütunlar için geçerlidir.

interpolate("linear") olması linear şekilde artacağı anlamına gelir.

```python
data2.resample("M").first().interpolate("linear")
```
