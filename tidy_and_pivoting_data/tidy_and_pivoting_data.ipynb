# **TIDY DATA**
*   Sütunlar değişkenleri, satırlar gözlemleri ifade eder ve gözlemler hücreleri oluşturur. Bir veri setinin temiz olarak nitelendirilmesi için bu şartları sağlaması gerekir.
*   Amaç veriyi analiz etmek için düzenli bir hale getirmektir.
*   Melt methodu ile veriler düzenlenir.
*   Melt methodu datadan istenilen featureları çıkararak düzenler.








```python
import pandas as pd 

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/pokemon.csv') #pokemon.csv datası dataframe'e çevrilir.
```

```python
data_new = data.head()  #Pokemon veri setinin ilk 5 örneği data_new değişkenine atanır.
data_new
```

melt methodu parametleri;
*   **frame** kısmına melt edilecek data yazılır.
*   **id_vars** datada değişmeden kalınması istenilen featuredır.
*   **value_vars** yeni oluşması istenilen variablelardır.






```python
melted = pd.melt(frame=data_new,id_vars = 'Name', value_vars= ['Speed','Defense']) #her bir pokemonun ismine göre melt işlemi yapılır.
melted                                                                              #Bulbasaur'ın Speed değişkeni 45'a eşitlendi.
                                                                                    #Ivysaur'un Defense değişkeni 63'e eşitlendi.
                                                                                #variable ve value isimleri default olarak atanır, istenirse değiştirilebilir.
```

# **PIVOTING DATA**

*   Pivoting ile melt edilen data tekrardan eski haline döndürülür.
*   **index** datada değişmesi istenilmeyen featuredır.
*   **columns** melt methodunda variable olarak kullanılan sütunlardır.
*   **values** melt methodunda kullanılan valuelardır.





```python
melted.pivot(index = 'Name', columns = 'variable',values='value')
```
