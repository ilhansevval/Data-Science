# **BUILDING DATA FRAMES FROM SCRATCH**
*   Dictionaryler de dataframelere dönüştürülebilir.
*   zip methodu ile tuple oluşturulur.
*   zip methodu ile indeks numaraları ve değerler bir arada tutulur.
*   Ziplenmiş bir data dictionary'ye çevrildiğinde indeks numaraları keys ve değerler de values olur.
*   **Broadcasting:** datada yeni bir feature oluşturup o feature'a yeni değerler atamaktır.








```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
country = ["Spain","France"] 
population = ["11","12"]     #country ve population adında listeler oluşturuldu
```

```python
list_label = ["country","population"]
list_col = [country,population]
zipped = list(zip(list_label,list_col))   #ziplenmiş datada listelerin adları list_label sütununa,listedeki veriler list_col sütununa atandı
```

```python
data_dict = dict(zipped)      #ziplenmiş data dictionary'e dönüştürüldü.
df = pd.DataFrame(data_dict)  #dictionary dataframe'e dönüştürüldü.
df
```

```python
df["capital"] = ["madrid","paris"] #dataframe'e yeni bir sütun eklenir ve bu yeni sütun en sona geçer.
df
```

```python
df["income"] = 0 #0 değeri tüm sütunlara atandığında bu broadcastingdir.
df
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/pokemon.csv') #pokemon.csv datası dataframe'e çevrilir.
```

```python
data.head()
```

```python
data1 = data.loc[:,["Attack","Defense","Speed"]]  #Attack,Defense ve Speed sütunları line plot ile çizdirilmiştir.
data1.plot()
plt.show()
```

```python
data1 = data.loc[:,["Attack","Defense","Speed"]] #grid ile grafik çizilirken dikdörtgen olarak ayırır, böylece incelemek kolaylaşır.
data1.plot(grid=True)
plt.show()
```

```python
data1 = data.loc[:,["Attack","Defense","Speed"]] 
data1.plot(grid=True, linestyle = ":") #linestyle grafiğin çizim şeklini değiştirilebilir.
plt.show()
```

```python
data1 = data.loc[:,["Attack","Defense","Speed"]] 
data1.plot(grid=True, alpha=0.5) #alpha ile çizgilerin saydamlığı artırılır, 0'a ne kadar yaklaşırsa o kadar artar.
plt.show()
```

```python
data1.plot(subplots = True,alpha=0.5) #subplot ile her bir feature ayrı ayrı çizilir.
plt.show()
```

```python
data1.plot(kind = "scatter",x="Attack",y = "Defense",color="red") #scatter plot iki feature karşılaştırmak için kullanılır.
plt.title("Scatter_Plot")
plt.show()
```

```python
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250)) #Defense sütunu 50 kutuyla 0-250 aralığında çizdirilir.
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1)
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250),normed = True,ax = axes[0])
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250),normed = True,ax = axes[1],cumulative = True)
plt.savefig('graph.png')
plt
```
