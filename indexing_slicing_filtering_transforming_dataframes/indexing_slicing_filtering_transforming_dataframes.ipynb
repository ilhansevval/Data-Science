# **SLICING DATA FRAME**
*   Sütun seçilirken tek köşeli parantez kullanılırken dataframe seçilirken 2 köşeli parantez kullanılır.



```python
print(type(data["HP"]))     
print(type(data[["HP"]]))
```

```python
data.loc[1:10,"HP":"Defense"] #satır için 1. ve 10. satır arasındaki,
                              #sütun için HP ve Defense sütunları arasındaki veriler yazılır.
                              #pandasta (inclusive,inclusive) olduğundan başlangıç ve bitişler dahildir.  
```

```python
data.loc[10:1:-1,"HP":"Defense"] # -1 ile tersten sıralama yapılır.
```

```python
data.loc[1:10,"Speed":] # ':' varsa başlangıç değerinden en sona kadar yazdırılır.
                        # ':' için herhangi bir başlangıç ve bitiş değeri belirtilmemişse baştan sona kadar yazdırır.
       #bu örnekte Speed sütunundan başlayarak dataframe'in sonundaki Legendary sütunu da dahil olmak üzere yazdırır.
```

```python
data.loc[1:10,:] #Sütun için başlangıç ve bitiş değerleri belirtilmediği için tüm sütunların ilk 10 satırını yazdırır.
```

# **FILTERING DATA FRAMES**


*   Bir boolean serisi yaratılır.
*   Boolean serisi True/False değerler içerir.
*   Yaratılan boolean serileri filtrelerdir.
*   Bu filtreler ile birbiri arasında karşılaştırma yapılır.





```python
boolean = data.HP > 200  #canı 200'den büyük olanları yazdırır.
data[boolean]
```

```python
first_filter = data.HP > 150      
second_filter = data.Speed > 35
data[first_filter & second_filter]  #burada iki filtre birleştirilir.
                                    #canı 150'den fazla olan ve hızı 35'ten fazla olanlar alınır.
```

```python
data.HP[data.Speed<15]   #HP sütunundaki Speed'i 15'ten küçük olanlar verilir.
```

# **TRANSFORMING DATA**


*   apply methodu kullanılır.
*   apply methodu ile yapılacak işlem seçilen verilerin hepsine teker teker uygulanılır.



```python
def div(n):
    return n/2      #div methodu ile veriler ikiye bölünür.
data.HP.apply(div)  #apply methodu ile 2'ye bölme işlemi HP sütunun
```

```python
data.HP.apply(lambda n : n/2) #lambda fonksiyon yazmanın hızlı yoludur.
```

```python
data["total_power"] = data.Attack + data.Defense
data.head()           #Attack ve Defense sütunundaki veriler toplanıp total_power adındaki sütuna bu veriler atanır.
```
