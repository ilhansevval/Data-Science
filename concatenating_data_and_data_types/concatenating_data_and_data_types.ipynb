# **CONCATENATING DATA**
*   **concat()** methodu ile iki dataframe birleştirilebilir.
*   **axis=0** dataframeleri dikey şekilde birleştirir.
*   **axis=1** dataframeleri yatay şekilde birleştirir.
*   **ignore_index =True** parametresi ile eski indeksler göz önünde bulundurulmaz, yeni indeksler atanır.






```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/pokemon.csv') #pokemon.csv datası dataframe'e çevrilir.
```

```python
data1 = data.head()  #data1 değişkenine ilk 5 satır data2 değişkenine son 5 satır atandı.
data2 = data.tail()
conc_data_row = pd.concat([data1,data2],axis =0,ignore_index =True)  #dataframeler dikey şekilde birleştirildi.
conc_data_row
```

```python
data1 = data['Attack'].head()
data2= data['Defense'].head()
conc_data_col = pd.concat([data1,data2],axis =1) #dataframeler yatay şekilde birleştirildi.
conc_data_col
```
